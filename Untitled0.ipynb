```python
def min_max_line(filename):
    with open(filename, 'r', encoding='utf-8') as file:
        lines = [line.rstrip('\n') for line in file.readlines()]

    # Ищем самую короткую строку (исключая возможные пустые строки)
    min_line = min([line for line in lines if line], key=len)

    # Ищем самую длинную строку
    max_line = max(lines, key=len)

    return f"Min = '{min_line}', max = '{max_line}'."
```
